```python
import numpy as np
#Пример срздания массивов:
lst1 = [1, 2, 3, 4, 5,6]
ar1 = np.array(lst1)
print(ar1)

matr = np.array([[1, 2, 3, 4], [1, 2, 3, 4]])
print(matr)

zeroes = np.zeros(5)
print("Массив нулей: ")
print(zeroes)

zeroesMatr = np.zeros((3, 4))
print("Двуменрный массив нулей размером 3 на 5")
print(zeroesMatr)

ones  = np.ones(5)
print("Массив единиц: ")
print(ones)

orange = np.arange(10)
print("Массив до 10: ")
print(orange)

orangeStep = np.arange(0, 10 , 2)
print("С шагом 2: ")
print(orangeStep)

#форма и тип массивов
print("Тип данных: ")
print(orangeStep.dtype)
print("Форма: ")
print(orangeStep.shape)

#Получение по индексу и срезы
print("Первый элемент и последний: ")
print(orangeStep[0], orangeStep[-1])
print("Срез: " )
print(orangeStep[2:4])

print("Срез для двумерного массива: ")
print(matr[0:2, :])

#Векторные операции и базовые статистики
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([10, 20, 30, 40, 50])

sm = arr2 + arr1 #Сложение
print("Сложение")
print(sm)

mult = arr1 * arr2
print("Умножение") #Умножение
print(mult)
#mean
print("Среднее по всем элементам: ")
print(np.mean(arr1))
```
